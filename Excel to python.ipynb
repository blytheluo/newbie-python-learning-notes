```python
#导入所需库
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
#读取数据
df = pd.read_excel('/Users/blytheluo/Downloads/示例数据.xlsx')
df.head()
```

```python
#数据筛选
df[df['薪资水平'] > 10000]
```

```python
#数据插入
classs = [0,5000,max(df['薪资水平'])]
group_names = ['低', '高']
df['newcol'] = pd.cut(df['薪资水平'], classs, labels=group_names)
df
```

```python
#删除列
del df['newcol']
df
```

```python
#数据排序
df1 = df
df1.sort_values('薪资水平', ascending=True)
```

```python
#处理缺失数据
df.isnull().sum()
#该数据中没有值为null的
```

```python
#数据去重
df.drop_duplicates(['创建时间'], inplace=True)
df
#会发现数据已经从825行变成了629行
```

```python
#数据格式修改
df1 = df
df1['创建时间'] = df1['创建时间'].dt.strftime('%Y-%m-%d')
df1
```

```python
#数据合并
df['合并列'] = df['地址'] + df['岗位']
df
```
